# Surprise Housing 

### Problem Statement

US based Surprise housing company is planning to enter the Australian housing market. The company wants to purchase houses below their actual prices and wants to flip it for a profit. This can be achieved through building a regression model that will predict the value of the prospective properties so a decision can be made to whether to invest in a given property or not.

### Encoding data

Let's encode MoSold to a cyclic value as months is cyclic in nature. We can do this by adding two additional variables cos_MoSold and sin_MoSold, as sin and cos are cyclic functions, we add both of these functions because if we add only either one of them, it leads to the same value occuring twice in the same cycle.

```python
import math

#converting MoSold variable in radians
df['MoSold_Norm'] = 2 * math.pi * df['MoSold'] / df['MoSold'].max()

#taking cosine of the radian values
df["cos_MoSold"] = np.cos(df["MoSold_Norm"])

#taking sine of the radian values
df["sin_MoSold"] = np.sin(df["MoSold_Norm"])
```

```python
df['sin_MoSold'].value_counts()
```

```python
#plotting cos and sin against each other to show they are cyclic
sns.scatterplot(df["cos_MoSold"], df["sin_MoSold"])
plt.show()
```

In the above graph each point refers to each month in an year, there are a total of 12 months in an year, by reading both the sine and cosine values together one is able to infer the cyclic nature of months.

```python
#let's drop MoSold and MoSold_Norm
df = df.drop(columns=['MoSold', 'MoSold_Norm'])
```

### Model Building and evaluation

#### Linear regression model

```python
# Instantiate
lm = LinearRegression()

# Fit a line
lm.fit(X_train, y_train)
```

```python
# Print the coefficients and intercept
print(lm.intercept_)
print(lm.coef_)
df_lr = pd.DataFrame(lm.coef_, index=X_train.columns, columns=['coefficient-value'])
```

```python
y_pred_train = lm.predict(X_train)
y_pred_test = lm.predict(X_test)

metric = []
#r2 score for train set
r2_train_lr = r2_score(y_train, y_pred_train)
print(r2_train_lr)
metric.append(r2_train_lr)

#r2 score for test set
r2_test_lr = r2_score(y_test, y_pred_test)
print(r2_test_lr)
metric.append(r2_test_lr)

#rss for train set
rss1_lr = np.sum(np.square(y_train - y_pred_train))
print(rss1_lr)
metric.append(rss1_lr)

#rss for test set
rss2_lr = np.sum(np.square(y_test - y_pred_test))
print(rss2_lr)
metric.append(rss2_lr)

#mse for train set
mse_train_lr = mean_squared_error(y_train, y_pred_train)
print(mse_train_lr)
metric.append(mse_train_lr**0.5)
#rmse for train set
print(mse_train_lr**0.5)

#mse for test set
mse_test_lr = mean_squared_error(y_test, y_pred_test)
print(mse_test_lr)
metric.append(mse_test_lr**0.5)
#rmse for test set
print(mse_test_lr**0.5)
```

Train r2 score value is very good at 95% whereas on test it is negative, this implies the model is heavily overfitted, and the model has memorised patterns from the data set and is not reliable for making predictions.

#### Ridge regression model

```python
# list of alphas to tune 
params = {'alpha': np.arange(0.001, 0.1, 0.001)}

lasso = Lasso()

# cross validation
model_cv = GridSearchCV(estimator = lasso, 
                        param_grid = params, 
                        scoring= 'r2', 
                        cv = folds, 
                        return_train_score=True,
                        verbose = 1)            

model_cv.fit(X_train, y_train) 
```

```python
# Lets calculate some metrics such as R2 score, RSS and RMSE

y_pred_train = lasso.predict(X_train)
y_pred_test = lasso.predict(X_test)

metric3 = []
r2_train_lr = r2_score(y_train, y_pred_train)
print(r2_train_lr)
metric3.append(r2_train_lr)

r2_test_lr = r2_score(y_test, y_pred_test)
print(r2_test_lr)
metric3.append(r2_test_lr)

rss1_lr = np.sum(np.square(y_train - y_pred_train))
print(rss1_lr)
metric3.append(rss1_lr)

rss2_lr = np.sum(np.square(y_test - y_pred_test))
print(rss2_lr)
metric3.append(rss2_lr)

mse_train_lr = mean_squared_error(y_train, y_pred_train)
print(mse_train_lr)
metric3.append(mse_train_lr**0.5)
print(mse_train_lr**0.5)

mse_test_lr = mean_squared_error(y_test, y_pred_test)
print(mse_test_lr)
metric3.append(mse_test_lr**0.5)
print(mse_test_lr**0.5)
```

The r2 score for the model is 85.6% for the training set and around 84.4% for the test dataset, the model is fitting the data very good as the test and train data are very close and can be used for making predictions.
